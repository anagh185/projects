```python
# Import package dependencies
import pandas as pd  # for dealing with data
import numpy as np
```

```python
df = pd.read_csv('Assignment6_data.csv')
df
```

```python
# Set the 'date' column as the index to help ensure dates show up in our plot.
df.set_index('Date', inplace=True)
df.head()
```

```python
df.info()
```

```python
returns_df = df.pct_change(1)
returns_df
```

```python
num_stocks = df.shape[1]
weights = [1/num_stocks] * num_stocks
weights
```

```python
vcv_matrix = returns_df.cov()
vcv_matrix
```

```python
var_p = np.dot(np.transpose(weights), np.dot(vcv_matrix, weights))
var_p
```

```python
sd_p = np.sqrt(var_p)
sd_p
```

```python
sdp_annual = sd_p * np.sqrt(250)
sdp_annual
```

```python
ind_risks = np.std(returns_df) * np.sqrt(250)
ind_risks
```

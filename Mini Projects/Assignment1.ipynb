```python
# Import package dependencies
import pandas as pd  # for dealing with data
import matplotlib.pyplot as plt  # for plotting
import seaborn as sns  # for making plots look nicer
sns.set()  # implementing Seaborn's style and themes
```

```python
# Load the Apple price data (Source: Yahoo! Finance)
df = pd.read_csv("/Users/anaghabhole/Downloads/AAPL.csv")  # Note that "../" changes the directory to the preceding folder.
```

```python
df.head()
```

```python
df = df[['Date', 'Adj Close']]
df.head()
```

```python
df.rename(columns={'Date' : 'date', 'Adj Close' : 'price_t'}, inplace=True)
df.head()
```

```python
df['daily_returns'] = df['price_t'].pct_change(1)
df.head()
```

```python
# Set the 'date' column as the index to help ensure dates show up in our plot.
df.set_index('date', inplace=True)
df.head()
```

```python
# Plot Apple's Price data
df['price_t'].plot(figsize=(12, 8))
```

```python
# Plot Apple's Returns data
df['daily_returns'].plot(figsize=(12, 8))
```
